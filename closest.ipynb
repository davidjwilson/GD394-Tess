```python
from lightkurve import search_targetpixelfile
pixelfile = search_targetpixelfile('GD394', sector=15).download(quality_bitmask='hardest')
```

```python
pixelfile.plot(frame=1,aperture_mask=pixelfile.pipeline_mask);
```

```python
pixelfile.interact()
```

```python
lc = lk.lightcurvefile.LightCurveFile(path='tess2019226182529-s0015-0000000259773610-0151-s_tp-custom-lc.fits') 
lc1 = fits.open('tess2019226182529-s0015-0000000259773610-0151-s_tp-custom-lc.fits')
data = lc1[1].data
lc2 = lk.LightCurve(time=data['TIME'], flux= data['FLUX'])
#lc2.scatter()

pg = lc2.to_periodogram(oversample_factor=10,minimum_period=0.8*u.day, maximum_period=1.5*u.day)
#plt.plot(pg.period, pg.power)

pg.plot()
print(pg.period_at_max_power)
```
