```python
import lightkurve as lk
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u 
import os
```

```python
sr = lk.search_targetpixelfile('GD394', mission='tess', sector=15)
tpf = sr.download()
```

```python
#just check we have the right lc
sr = lk.search_lightcurvefile('GD394', mission='tess', sector=15)
lcf = sr.download()
```

```python
cutout_size = 50
cutout = lk.search_tesscut('{} {}'.format(tpf.ra, tpf.dec))[1].download(cutout_size=(cutout_size, cutout_size))
```

```python
ax = cutout.plot()
ax.set_title('')
```

```python
cutout.to_lightcurve(aperture_mask='all').plot();
```

```python
# Make a thumbnail of the tpf to find the pixels with the lowest flux
thumb = np.nanmedian(cutout.flux, axis=0)
bkg_aper = thumb < np.nanpercentile(thumb, 10)

# Take the median of these pixels in time
bkg = np.median(cutout.flux[:, bkg_aper], axis=1)
```

```python
from photutils import DAOStarFinder
daofind = DAOStarFinder(fwhm=1.0, threshold=150)  
sources = daofind(np.median(cutout.flux, axis=0)) 
xs, ys = np.asarray(sources.to_pandas()[['xcentroid', 'ycentroid']]).T
```

```python
xs, ys
```

```python
apers = {}
lcs = {}
positions = {}
for idx, x, y in zip(range(len(xs)), xs, ys):
    aper = cutout.create_threshold_mask(5, (x, y))
    # No tiny apertures
    if aper.sum() < 3:
        continue
    # No previously found apertures
    if any([(aper == a).all() for k, a in apers.items()]):
        continue
    lc = (cutout.to_lightcurve(aperture_mask=aper) - (aper.sum()*bkg))
    lcs[idx] = lc
    apers[idx] = aper
    positions[idx] = (x, y)
    #lc = lc.flatten(10001)#.plot(ax=ax, c='r', alpha=0.2)
    pg = lc.to_periodogram(oversample_factor=10,minimum_period=0.5*u.day, maximum_period=2*u.day)
    #pg.plot()
    
    
    # Plotting
    fig = plt.figure(figsize=(12, 4))
    ax = plt.subplot2grid((1, 4), (0, 0), colspan=3)
    #lc.plot(ax=ax)
    pg.plot(ax=ax)
    ax = plt.subplot2grid((1, 4), (0, 3))
    cutout.plot(ax=ax, aperture_mask=aper, show_colorbar=False)
    ax.set_title('')
    ax.set_xlabel('')
    ax.set_ylabel('')  
```

```python
# Find the aperture closest to the center (which is the target)
d = [np.sum((np.asarray(p) - cutout_size/2)**2)**0.5 for k, p in positions.items()]
target = list(positions.keys())[np.argmin(d)]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
for k in lcs.keys():
    if k == target:
        continue
    if k == list(lcs.keys())[0]:
        lcs[k].flatten(301).plot(ax=ax, c='r', alpha=0.2, label='{} Nearest Neighbors'.format(len(lcs) - 1))
    else:
        lcs[k].flatten(301).plot(ax=ax, c='r', alpha=0.2)
lcs[target].flatten(301).plot(ax=ax, c='k', label='GD394')
plt.ylim(0.985, 1.01)
plt.legend()
#plt.savefig('NN.png', dpi=200, bbox_inches='tight')
```
