```python
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams["figure.figsize"] = (14, 5)
rcParams["font.size"] = 20
```

Jean Dupuis has sent some of the original lightcurves for gd394. Let's see if they are useful.  

```python
import os
lcs = os.listdir('.')
print(lcs)
```

All of the lightcuves are in time (s), counts (/s?), error. ds_5760_ltc_obs2.ascii is the longest. Let's plot them all first.

```python
lcs = ['gd394_sca_alc_oct92', 'gd394_scb_lex_oct92', 'ds_5760_ltc_obs2.ascii', 'gd394_sca_lex_nov96', 'gd394_scb_alc_nov96']
#clear out the notebook files
for lc in lcs:
    plt.figure(lc)
    t, c, e = np.loadtxt(lc, unpack=True)
    plt.errorbar(t/86400., c, yerr=e, marker='o', ls='none')
    plt.title(lc.replace('_',' '))
    plt.xlabel('Time (d)')
    plt.ylabel('Counts (s$^{-1}$)')
```

That's wonderful, so nice to see that these things are real!

```python
#Lomb-Scargles
from astropy.timeseries import LombScargle
periods = []
for lc in lcs:
    plt.figure(lc)
    t, c, e = np.loadtxt(lc, unpack=True)
    t, c, e = t[c>0], c[c>0], e[e>0]
    t = t/86400. #days
    frequency, power = LombScargle(t, c, e).autopower(minimum_frequency=1/5., maximum_frequency=1/0.5, samples_per_peak=100)
    period = 1 / frequency[np.argmax(power)]
    plt.plot(1 / frequency, power)
    plt.xlabel("Period (days)")
    plt.ylabel("Power")
    plt.title(lc.replace('_',' '))
    print(lc,period)
    periods.append(period)
```

```python
print(periods)
```

So the 92 data is probably useless, but all the others- hello 1.15 day period! Let's fold them all on that and see.

I'm also going to comment out that clearly bad pixel in the ds file and run everythig again without it

```python
av_period = np.mean(periods[2:])
print(av_period) 
```

Find the t_0 from the earliest lightcurve:

```python
t_s = []
for lc in lcs:
    t, c, e = np.loadtxt(lc, unpack=True)
    t_s.append(t[0])
t_0 = min(t_s)
print(t_0)
```

```python
for lc in lcs:
    plt.figure(lc)
    t, c, e = np.loadtxt(lc, unpack=True)
    t = (t-t_0)/86400.
    fold_t = (t/av_period) % 1
    plt.errorbar(fold_t, c, yerr=e, marker='o', ls='none')
    plt.title(lc.replace('_',' '))
    plt.xlabel('Phase')
    plt.ylabel('Counts (s$^{-1}$)')
```

And in normalised counts:

```python
for lc in lcs:
    plt.figure(lc)
    t, c, e = np.loadtxt(lc, unpack=True)
    t = (t-t_0)/86400.
    fold_t = (t/av_period) % 1
    plt.errorbar(fold_t, c-np.median(c), yerr=e, marker='o', ls='none')
    plt.title(lc.replace('_',' '))
    plt.xlabel('Phase')
    plt.ylabel('Normalised Counts')
```

Looks like the uncertainty in the period is too large to get an ephemeris. Let's try sticking all of the normalised data together...

```python
t_all, c_all, e_all = np.array([], dtype=float), np.array([], dtype=float), np.array([], dtype=float)

for lc in lcs:
    t, c, e = np.loadtxt(lc, unpack=True)
    t, c, e = t[c>0], c[c>0], e[e>0]
    cn = c/np.median(c)
    t_all = np.concatenate((t_all, t))
    c_all = np.concatenate((c_all, cn))
    e_all = np.concatenate((e_all, e))

plt.errorbar(t_all, c_all, e_all, marker='o', ls='none')

args = np.argsort(t_all)
t_all= t_all[args]
c_all = c_all[args]
e_all= e_all[args]
```

```python
frequency, power = LombScargle(t_all/86400., c_all, e_all).autopower(minimum_frequency=1/3., maximum_frequency=1/0.5, samples_per_peak=100)
period = 1 / frequency[np.argmax(power)]
plt.plot(1 / frequency, power)
plt.xlabel("Period (days)")
plt.ylabel("Power")
print(period)
```
